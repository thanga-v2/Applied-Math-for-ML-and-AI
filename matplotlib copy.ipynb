```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.plot(x-axis,y-axis)

plt.plot(2,4)
```

```python
# r is for red
plt.plot(2,4,'r*')
```

```python
# x-axis
x = np.arange(-8,9)

y = x ** 2

print ('x is a vector \n',x)
print ('y is a vector \n ',y)
# y = x * 2
# y = x + 2
print('\n x and y changing in a space \n')
plt.plot(x,y,'r')
# gs -> green squares
plt.plot(x,y/2,'gs') 
```

```python
# ploting lines
# plt.plot([x-axis],[y-axis])

plt.plot([0,3],[0,5])
```

```python
# ploting lines
# plt.plot([x-axis],[y-axis])

plt.plot([0,3],[0,5])
plt.plot([-4,1],[-3,2])
```

```python
# ploting lines
# plt.plot([x-axis],[y-axis])

plt.plot([0,3],[0,5],label='first line')
plt.plot([-4,1],[-3,2], label = 'second line')
plt.legend()
```
