```python



# orthogonal

# orthoganality -> two vectors, in which its dot product result is zero

v1 = [ 2, 4, -1]

v2 = [-2, 1, 0]

np.dot(v1,v2)

#  #  #  v1 is orthogonal to v2
```

```python

# 3 2d vectors where two of them are orthogonal

v1 = np.array([2,4])
v2 = np.array([8,2])
v3 = np.array([-4,2])

np.dot(v1,v2)
```

```python
np.dot(v2,v3)
```

```python
np.dot(v1,v3)

# v1 and v3 are orthogonal
```

```python
# plot all the three in a 2d plane



plt.plot((0,v1[0]),(0,v1[1]),'r--',label="v1")
plt.plot((0,v2[0]),(0,v2[1]),'g--',label="v2")
plt.plot((0,v3[0]),(0,v3[1]),'b--',label="v3")
plt.xlim(-6,6)
plt.ylim(-6,6)

# setting aspect equal 

plt.gca().set_aspect('equal')
plt.legend()


```
